## 問題 1

**問題**: Autograd `Tensor` `__sub__`, `__truediv__`, `matmul` .

### 厳密な解説

制御する要因: **operation: subtraction, division, and matrix multiplication**。  測定指標: **autograd gradients compared with closed-form references**。  解釈と限界: The local Jacobians are 1/-1, reciprocal/quotient, and matrix transposes. PyTorch supplies an independent reverse-mode reference for the requested operators.


```python
import torch
a=torch.tensor([[2.,4.]],requires_grad=True); b=torch.tensor([[1.,2.]],requires_grad=True); W=torch.tensor([[1.,2.],[3.,4.]],requires_grad=True); z=((a-b)/b)@W; z.sum().backward(); print({'a_grad':a.grad.tolist(),'b_grad':b.grad.tolist(),'W_grad':W.grad.tolist()}); assert a.grad.shape==a.shape and W.grad.shape==W.shape
```

## 問題 2

**問題**: Autograd $f(x, y) = (x + y)^2 - xy$ $\partial f/\partial x, \partial f/\partial y$ 計算 数値微分 検証.

### 厳密な解説

制御する要因: **derivative method: analytic versus central difference**。  測定指標: **absolute derivative error for x and y**。  解釈と限界: For f=(x+y)^2-xy, derivatives are 2(x+y)-y and 2(x+y)-x. Central differences verify both independently.


```python
import torch
x,y=1.3,-.4; f=lambda a,b:(a+b)**2-a*b; e=1e-6; ana=[2*(x+y)-y,2*(x+y)-x]; num=[(f(x+e,y)-f(x-e,y))/(2*e),(f(x,y+e)-f(x,y-e))/(2*e)]; err=max(abs(a-b) for a,b in zip(ana,num)); print({'analytic':ana,'numeric':num,'max_error':err}); assert err<1e-8
```

## 問題 3

**問題**: PyTorch 3 MLP , バックプロパゲーション `W1.grad`, `W2.grad` 出力.

### 厳密な解説

制御する要因: **one backward pass through a three-layer MLP**。  測定指標: **W1 and W2 gradient shapes and norms**。  解釈と限界: The chain rule produces gradients matching each weight matrix. Nonzero finite norms demonstrate that the loss reached both layers.


```python
import torch
torch.manual_seed(70); W1=torch.randn(4,6,requires_grad=True); W2=torch.randn(6,3,requires_grad=True); X=torch.randn(8,4); y=torch.randint(0,3,(8,)); loss=torch.nn.functional.cross_entropy(torch.relu(X@W1)@W2,y); loss.backward(); print({'W1_shape':list(W1.grad.shape),'W1_norm':W1.grad.norm().item(),'W2_shape':list(W2.grad.shape),'W2_norm':W2.grad.norm().item()}); assert torch.isfinite(W1.grad).all()
```

## 問題 4

**問題**: 1, 5, 10, 20 MLP gradient norm 比較 .

### 厳密な解説

制御する要因: **depth: 1, 5, 10, 20**。  測定指標: **input gradient norm**。  解釈と限界: All layers use the same scalar contraction 0.8, so the chain rule predicts 0.8^depth exactly. This controlled construction isolates vanishing gradients rather than mixing in random matrices.


```python
import torch
for depth in (1,5,10,20):
 x=torch.tensor(1.,requires_grad=True); h=x
 for _ in range(depth): h=.8*h
 h.backward(); print({'depth':depth,'input_grad_norm':abs(x.grad.item()),'reference':.8**depth}); assert abs(x.grad.item()-.8**depth)<1e-6
```

## 問題 5

**問題**: He ( `randn`) 比較 MLP 学習 .

### 厳密な解説

制御する要因: **initialization: He scaling versus unscaled randn**。  測定指標: **activation variance and input-gradient norm after 20 ReLU layers**。  解釈と限界: Matched Gaussian draws isolate the scaling factor. Unscaled weights cause exploding variance; He scaling targets stable second moments. This measures signal propagation, not full-task accuracy.


```python
import torch
for name,scale in [('he',(2/64)**.5),('randn',1.)]:
 torch.manual_seed(71); x=torch.randn(128,64,requires_grad=True); h=x
 for _ in range(20): h=torch.relu(h@(torch.randn(64,64)*scale))
 h.square().mean().backward(); print({'init':name,'activation_variance':h.var().item(),'input_grad_norm':x.grad.norm().item()})
```
