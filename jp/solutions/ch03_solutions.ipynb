## 問題 1

**問題**: f(x) = x^3 - 3x^2 + 2 導関数 数値微分 検証


```python
import numpy as np

def f(x): return x**3 - 3*x**2 + 2
def f_prime(x): return 3*x**2 - 6*x  # 解析的

def num_diff(f, x, h=1e-5):
    return (f(x+h) - f(x-h)) / (2*h)

x0 = 1.5
print(f"f({x0}) = {f(x0)}")
print(f"f'({x0}) 解析的 = {f_prime(x0)}")
print(f"f'({x0})  = {num_diff(f, x0):.6f}")
print(f"誤差 = {abs(f_prime(x0) - num_diff(f, x0)):.2e}")
```
