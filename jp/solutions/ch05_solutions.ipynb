## 問題 1

**問題**: 4テキスト/6テキスト テキスト テキスト 比較


```python
import numpy as np
def entropy(p):
    p = np.array(p, dtype=float)
    p = p[p > 0]
    return -np.sum(p * np.log(p))

H_4 = entropy([0.25]*4)
H_6 = entropy([1/6]*6)
print(f"4テキスト テキスト H = {H_4:.4f} nat = {H_4/np.log(2):.4f} bit")
print(f"6テキスト テキスト H = {H_6:.4f} nat = {H_6/np.log(2):.4f} bit")
print("6テキスト テキスト テキスト テキスト.")
```
