## Problem 1

**Problem**: 3Dimension Vector a=[1,2,3], b=[4,-1,0]text Dot Product, L2 Norm, Cosine Similarity


```python
import numpy as np
a = np.array([1, 2, 3], dtype=float)
b = np.array([4, -1, 0], dtype=float)

# Dot Product
dot = np.dot(a, b)
print(f"a · b = {a[0]}*{b[0]} + {a[1]}*{b[1]} + {a[2]}*{b[2]} = {dot}")

# L2 Norm
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
print(f"|a| = sqrt({a[0]**2}+{a[1]**2}+{a[2]**2}) = {norm_a:.4f}")
print(f"|b| = sqrt({b[0]**2}+{b[1]**2}+{b[2]**2}) = {norm_b:.4f}")

# Cosine Similarity
cos_sim = dot / (norm_a * norm_b)
print(f"cos(a, b) = {dot}/{norm_a:.4f}*{norm_b:.4f} = {cos_sim:.4f}")
print(f"Angle = {np.degrees(np.arccos(cos_sim)):.2f}degrees")
```
