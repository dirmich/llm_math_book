**Problem**: ABtext BA Calculation (Matrix Multiplication text text text text)


```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

AB = A @ B
BA = B @ A
print(f"A =\n{A}")
print(f"B =\n{B}")
print(f"AB =\n{AB}")
print(f"BA =\n{BA}")
print(f"\nAB == BA? {np.array_equal(AB, BA)}")
print("=> Matrix Multiplicationtext text text text!")
```
