## Problem 1

**Problem**: Apply your own BPE implementation to Korean text and describe the challenges encountered.

### Rigorous solution

Controlled factor: **BPE merge count on a fixed Korean corpus**.  Reported metric: **token count and learned merge sequence**.  Interpretation and limitation: BPE repeatedly merges the most frequent adjacent symbol pair. Korean syllable/morpheme variation fragments counts; ties also make results corpus/order dependent.


```python
import torch
from collections import Counter
words=[list(w) for w in '낮 말은 새가 듣고 밤 말은 쥐가 듣는다'.split()]
for step in range(5):
 pairs=Counter((a,b) for w in words for a,b in zip(w,w[1:])); pair=max(pairs,key=lambda p:(pairs[p],p)); merged=''.join(pair); words=[[merged if i<len(w)-1 and (w[i],w[i+1])==pair else w[i] for i in range(len(w)) if i==0 or (w[i-1],w[i])!=pair] for w in words]; print({'step':step+1,'merge':pair,'tokens':sum(map(len,words))})
```

## Problem 2

**Problem**: Train BPE and WordPiece on the same corpus and compare their merge orders.

### Rigorous solution

Controlled factor: **selection score: BPE pair frequency versus WordPiece association score**.  Reported metric: **first selected pair**.  Interpretation and limitation: BPE favors raw frequent pairs; WordPiece-style scoring divides by unigram frequencies and can favor a more exclusive pair. Same corpus makes the objective the only changed factor.


```python
import torch
from collections import Counter
seq=[tuple(w) for w in 'low lower newest widest low low'.split()]; pairs=Counter((a,b) for w in seq for a,b in zip(w,w[1:])); uni=Counter(c for w in seq for c in w); bpe=max(pairs,key=lambda p:pairs[p]); wp=max(pairs,key=lambda p:pairs[p]/(uni[p[0]]*uni[p[1]])); print({'BPE_first':bpe,'WordPiece_first':wp,'BPE_count':pairs[bpe],'WordPiece_score':pairs[wp]/(uni[wp[0]]*uni[wp[1]])})
```

## Problem 3

**Problem**: Compare the average number of tokens for vocabulary sizes of 100, 500, and 1000.

### Rigorous solution

Controlled factor: **BPE vocabulary budget: 100, 500, 1000 (capped by available pairs)**.  Reported metric: **mean tokens per sentence**.  Interpretation and limitation: The same local corpus is tokenized after increasing merge budgets. A tiny corpus exhausts useful pairs before large nominal vocabulary sizes, which is an explicit limitation.


```python
import torch
from collections import Counter
base=['the quick brown fox','the quicker brown fox','quick foxes jump']*20
for budget in (100,500,1000):
 words=[list(w) for s in base for w in s.split()]; merges=min(budget-30,40)
 for _ in range(merges):
  pairs=Counter((a,b) for w in words for a,b in zip(w,w[1:]));
  if not pairs: break
  p=max(pairs,key=pairs.get); m=''.join(p); words=[[m if i<len(w)-1 and (w[i],w[i+1])==p else w[i] for i in range(len(w)) if i==0 or (w[i-1],w[i])!=p] for w in words]
 print({'vocab_budget':budget,'mean_tokens_per_sentence':sum(map(len,words))/len(base)})
```

## Problem 4

**Problem**: Explain why Byte-level BPE eliminates the UNK token.

### Rigorous solution

Controlled factor: **input Unicode strings**.  Reported metric: **UTF-8 byte round-trip failures and byte vocabulary bound**.  Interpretation and limitation: Every Unicode string maps to bytes in 0..255, so a byte-level base vocabulary has no UNK. Merges only combine known byte sequences and cannot remove coverage.


```python
import torch
samples=['한글','🙂','𝛑','mixed언어']; rows=[]
for s in samples: ids=list(s.encode('utf-8')); rows.append((s,ids,bytes(ids).decode('utf-8')==s))
print({'base_vocabulary_size':256,'round_trips':rows}); assert all(ok for _,_,ok in rows)
```
