## Problem 1

**Problem**: Apply SGD, Momentum, and Adam to the Rosenbrock function and compare convergence paths.

### Rigorous solution

Controlled factor: **optimizer: SGD, Momentum, Adam**.  Reported metric: **final Rosenbrock objective and distance to optimum**.  Interpretation and limitation: Each optimizer starts at the same point and receives the exact analytic gradient for the same step budget. Learning rates are named protocol choices, so trajectory rankings are not claimed universal.


```python
import torch
def fg(x):
 a,b=x; return (1-a)**2+100*(b-a*a)**2, torch.stack((-2*(1-a)-400*a*(b-a*a),200*(b-a*a)))
for name in ('sgd','momentum','adam'):
 x=torch.tensor([-1.2,1.]); m=torch.zeros(2); v=torch.zeros(2)
 for t in range(1,2001):
  loss,g=fg(x); lr=1e-3 if name!='adam' else 2e-3
  if name=='sgd': step=g
  elif name=='momentum': m=.9*m+g; step=m
  else: m=.9*m+.1*g; v=.999*v+.001*g*g; step=(m/(1-.9**t))/(torch.sqrt(v/(1-.999**t))+1e-8)
  x=x-lr*step
 print({'optimizer':name,'loss':loss.item(),'distance':torch.linalg.vector_norm(x-torch.ones(2)).item()})
```

## Problem 2

**Problem**: Show why Adam's bias correction is needed by comparing the magnitude of $\mathbf{m}_t$ in early steps.

### Rigorous solution

Controlled factor: **Adam first-moment correction: raw versus bias-corrected**.  Reported metric: **moment estimate during first five steps**.  Interpretation and limitation: For constant gradient g, m_t=(1-beta^t)g; division by 1-beta^t removes initialization bias exactly.


```python
import torch
g=2.; beta=.9; m=0.
for t in range(1,6): m=beta*m+(1-beta)*g; corrected=m/(1-beta**t); print({'step':t,'raw_m':m,'corrected_m':corrected}); assert abs(corrected-g)<1e-12
```

## Problem 3

**Problem**: Compare Adam vs AdamW on MNIST with weight decay set to 0.01.

### Rigorous solution

Controlled factor: **regularization rule: Adam L2 coupling versus AdamW decoupling**.  Reported metric: **parameter value after five zero-loss-gradient steps**.  Interpretation and limitation: With zero data gradient, coupled Adam normalizes the L2 term through moments, whereas AdamW multiplies weights by 1-lr*wd. This isolates the mathematical difference, not MNIST accuracy.


```python
import torch
for name in ('adam_l2','adamw'):
 w=torch.tensor(1.); m=v=0.; lr=.1; wd=.01
 for t in range(1,6):
  if name=='adam_l2': g=wd*w; m=.9*m+.1*g; v=.999*v+.001*g*g; w-=lr*(m/(1-.9**t))/(v/(1-.999**t))**.5
  else: w*=1-lr*wd
 print({'rule':name,'weight_after_5_steps':w.item()})
```

## Problem 4

**Problem**: Run Adam with learning rates $10^{-2}, 10^{-3}, 10^{-4}$ and compare convergence speed.

### Rigorous solution

Controlled factor: **Adam learning rate: 1e-2, 1e-3, 1e-4**.  Reported metric: **quadratic loss after 100 matched steps**.  Interpretation and limitation: The objective, start, beta values, and steps are fixed; only learning rate changes. The result describes this finite budget, not an unconditional best rate.


```python
import torch
for lr in (1e-2,1e-3,1e-4):
 x=torch.tensor(1.); m=v=0.
 for t in range(1,101): g=2*x; m=.9*m+.1*g; v=.999*v+.001*g*g; x-=lr*(m/(1-.9**t))/(torch.sqrt(v/(1-.999**t))+1e-8)
 print({'learning_rate':lr,'loss_after_100':x.square().item()})
```

## Problem 5

**Problem**: Explain why LLM training becomes unstable without a warmup phase.

### Rigorous solution

Controlled factor: **schedule: immediate full rate versus linear warmup**.  Reported metric: **maximum parameter update over the first ten steps**.  Interpretation and limitation: Early adaptive statistics and large model gradients can make immediate full-rate updates abrupt. Warmup bounds early update magnitude; this deterministic sequence demonstrates that mechanism.


```python
import torch
grads=torch.tensor([20.,12.,8.,5.,3.,2.,1.,1.,1.,1.])
for mode in ('no_warmup','linear_warmup'):
 w=torch.tensor(0.); updates=[]
 for t,g in enumerate(grads,1): lr=.1 if mode=='no_warmup' else .1*min(1,t/5); u=lr*g; w-=u; updates.append(abs(float(u)))
 print({'schedule':mode,'max_early_update':max(updates),'parameter':w.item()})
```
