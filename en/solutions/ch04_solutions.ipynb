## Problem 3

**Problem**: text text: text text text


```python
# P(text)=0.3, P(free|text)=0.8, P(free|text)=0.05
P_S = 0.3; P_NS = 0.7
P_F_given_S = 0.8; P_F_given_NS = 0.05
P_F = P_F_given_S * P_S + P_F_given_NS * P_NS
P_S_given_F = P_F_given_S * P_S / P_F
print(f"P(text|freetext) = {P_S_given_F:.4f}")
print(f"=> freetext text text text text {P_S_given_F*100:.1f}%")
```
