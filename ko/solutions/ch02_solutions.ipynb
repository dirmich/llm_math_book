## 문제 1

**문제**: AB와 BA 계산 (행렬곱 교환법칙 성립 안 함)


```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

AB = A @ B
BA = B @ A
print(f"A =\n{A}")
print(f"B =\n{B}")
print(f"AB =\n{AB}")
print(f"BA =\n{BA}")
print(f"\nAB == BA? {np.array_equal(AB, BA)}")
print("=> 행렬곱은 교환법칙이 성립하지 않는다!")
```
