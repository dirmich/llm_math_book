## 문제 1

**문제**: f(x) = x^3 - 3x^2 + 2의 도함수와 수치 미분 검증


```python
import numpy as np

def f(x): return x**3 - 3*x**2 + 2
def f_prime(x): return 3*x**2 - 6*x  # 해석적

def num_diff(f, x, h=1e-5):
    return (f(x+h) - f(x-h)) / (2*h)

x0 = 1.5
print(f"f({x0}) = {f(x0)}")
print(f"f'({x0}) 해석적 = {f_prime(x0)}")
print(f"f'({x0}) 수치 = {num_diff(f, x0):.6f}")
print(f"오차 = {abs(f_prime(x0) - num_diff(f, x0)):.2e}")
```
