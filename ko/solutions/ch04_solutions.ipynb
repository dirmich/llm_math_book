## 문제 3

**문제**: 베이즈 정리: 스팸 메일 확률


```python
# P(스팸)=0.3, P(free|스팸)=0.8, P(free|정상)=0.05
P_S = 0.3; P_NS = 0.7
P_F_given_S = 0.8; P_F_given_NS = 0.05
P_F = P_F_given_S * P_S + P_F_given_NS * P_NS
P_S_given_F = P_F_given_S * P_S / P_F
print(f"P(스팸|free포함) = {P_S_given_F:.4f}")
print(f"=> free가 포함된 메일이 스팸일 확률은 {P_S_given_F*100:.1f}%")
```
